# GDP per capita

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data.groupby(by=['Region'])['GDP per capita (current US$)'].mean().sort_values().plot(kind='bar')
plt.title("Mean GDP per capita")
plt.show()
```

Northern America has highest GDP.

# Intenational trade:imports

```python
data.groupby(by='Region')['International trade: Imports (million US$)'].mean().sort_values().plot(kind ='bar')
plt.title("Mean International trade: Imports (million US$) ")
plt.show()
```

```python
data.groupby(by='Region')['GDP per capita (current US$)'].std().sort_values().plot(kind ='bar')
plt.title("std of GDP per capita (current US$)")
plt.show()
```
